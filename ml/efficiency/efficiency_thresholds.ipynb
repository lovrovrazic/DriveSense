```python
import numpy as np
import matplotlib.pyplot as plt
```

### Normal driving

```python
for e in ["braking", "acceleration", "steering"]:
    result = np.array([])
    print("{:>20s} | {:s}".format("events", e))
    for name in [ "normal1", "normal2", "normal3"]:
        file_name = "data_events/" + name + "_" + e + ".npy"
        event = np.load(file_name)
        print("{:>20s} | {:d}".format(name, len(event)))
        result = np.concatenate((result, event))
    print("{:>20s} | {:d}".format("number of events", len(result)))
    print("{:>20s} | {:f}".format("80 percentile", np.percentile(result, 80)))
    plt.title(e)
    plt.hist(result, alpha=0.5, histtype='bar', ec='black',bins=50)
    plt.ylabel('number of events')
    plt.xlabel('peaks')
    plt.show()
    
```

### Agressive driving

```python
for e in ["braking", "acceleration", "steering"]:
    result = np.array([])
    print("{:>20s} | {:s}".format("events", e))
    for name in ["aggressive1", "aggressive2", "aggressive3"]:
        file_name = "data_events/" + name + "_" + e + ".npy"
        event = np.load(file_name)
        print("{:>20s} | {:d}".format(name, len(event)))
        result = np.concatenate((result, event))
    print("{:>20s} | {:d}".format("number of events", len(result)))
    print("{:>20s} | {:f}".format("80 percentile", np.percentile(result, 80)))
    plt.title(e)
    plt.hist(result, alpha=0.5, histtype='bar', ec='black',bins=50)
    plt.ylabel('number of events')
    plt.xlabel('peaks')
    plt.show()
```
